# アリゴリズム 5.2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# アルゴリズム 5.1
def OT1D(a, b, x, y):
    asx = np.argsort(x)
    asy = np.argsort(y)
    x = x[asx]
    a = a[asx]
    y = y[asy]
    b = b[asy]
    r = 0
    i = 0
    j = 0
    while i < len(x) or j < len(y):
        p = min(a[i], b[j])
        r += p * np.abs(x[i] - y[j])
        a[i] -= p
        b[j] -= p
        if a[i] == 0:
            i += 1
        if b[j] == 0:
            j += 1
    return r
```

```python
# アルゴリズム 5.2
def SW(a, b, x, y, K=100):
    r = 0
    for k in range(K):
        w = np.random.randn(x.shape[1])
        w = w / np.sqrt((w ** 2).sum())
        px = x @ w
        py = y @ w
        r += OT1D(a, b, px, py)
    return r / K
```

```python
# データ生成
np.random.seed(0)
n = 1000
x = np.random.randn(n, 2) # N(0, I)
y = np.random.randn(n, 2) # N(0, I)
a = np.ones(n) / n
b = np.ones(n) / n
plt.scatter(x[:, 0], x[:, 1])
plt.scatter(y[:, 0], y[:, 1])
```

```python
SW(a, b, x, y)
```

```python
# データ生成
np.random.seed(0)
n = 1000
x = np.random.randn(n, 2) + 1 # N(1, I)
y = np.random.randn(n, 2) # N(0, I)
a = np.ones(n) / n
b = np.ones(n) / n
plt.scatter(x[:, 0], x[:, 1])
plt.scatter(y[:, 0], y[:, 1])
```

```python
SW(a, b, x, y)
```
